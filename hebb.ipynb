```python
import numpy as np
import matplotlib.pyplot as plt

from hebb import Neuron, NeuralNetwork
from util import show_history, animate, plot_connections
```

```python
network = NeuralNetwork(size=100, input_size=5, threshold=1.0)
network.connect_randomly(connection_probability=0.06, weight_range=(0.18, 0.23)) #(0.02, 0.1): very weak activations
plot_connections(network)
```

```python
input_neuron_history, neuron_history = network.simulate(iterations=500)
```

```python
show_history(network, input_neuron_history, neuron_history)
```

```python
### Save the neurons
saved_input_neurons = (network.input_neurons).copy()
saved_neurons = (network.neurons).copy()
saved_input_positions = (network.input_positions).copy()
saved_positions = (network.positions).copy()

### Reset each neuron's potential
for input_neuron in saved_input_neurons:
    input_neuron.potential = 0.0
    input_neuron.hebb = 0.00
for neuron in saved_neurons:
    neuron.potential = 0.0
    neuron.hebb = 0.0
```

```python
new_input_neurons = saved_input_neurons.copy()
new_neurons = saved_neurons.copy()

### Update hebbian learning rate for new neurons
for neuron in new_input_neurons:
    neuron.hebb = 0.1

for neuron in new_input_neurons:
    neuron.hebb = 0.1
```

```python
network = NeuralNetwork(size=20, input_size=5, threshold=1.0)

# Load in old connections
network.input_neurons = new_input_neurons
network.neurons = new_neurons
network.input_positions = saved_input_positions
network.positions = saved_positions

# Plot initial connections
plot_connections(network)

# Run the simulation, this time with hebbian learning
input_neuron_history, neuron_history = network.simulate(iterations=1000)
# animate(network, input_neuron_history, neuron_history)

# Plot final connections
plot_connections(network)
```

```python
show_history(network, input_neuron_history, neuron_history)
```

```python
animate(network, input_neuron_history, neuron_history)
```
